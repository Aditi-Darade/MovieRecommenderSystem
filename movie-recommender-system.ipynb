```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits,on='title') #Merged the two (credits and movies) data frames on the column "titles"
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']] #onl keep these columns from the dataset and delete the rest
```

```python
movies.isnull().sum() #check if any data is null
```

```python
movies.dropna(inplace=True) #drop all the null data
```

```python
movies.duplicated().sum() #check if any data is duplicated
```

```python
movies.iloc[0].genres #find datatype of genres
```

```python
import ast 
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# convert datatype of genres from '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# to ['Action','Adventure','Fantasy','Science Fiction']
def convert(text):
    L = []    #Creating a list
    for i in ast.literal_eval(text):   #Converting strings to list
        L.append(i['name']) 
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
#kepping only the first three cast members
def convert3(text):
    L = []    #Creating a list
    counter = 0
    for i in ast.literal_eval(text):   #Converting strings to list
        if counter !=3:
            L.append(i['name']) 
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(text):
    L = []    #Creating a list
    for i in ast.literal_eval(text):   #Converting strings to list
        if i['job'] == 'Director':
            L.append(i['name']) 
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
#convert string to list 
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x] ) #remove space in genres, eg Science Fiction will be ScienceFiction
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x] )
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x] )
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x] )
movies.head()
```

```python
#create a new column called tags from the movies dataset which will be a concatination of overview, genres, keywords, cast, crew 
movies ['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
#create a new datadframe with movie_id, title, and tags column
new_df = movies[['movie_id','title', 'tags']]
new_df
```

```python
#convert list to string
new_df ['tags'].apply(lambda x: " ".join(x))
```

```python
new_df ['tags'] = new_df ['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df ['tags'][0]
```

```python
#convert tags content to lowercase
new_df ['tags'].apply(lambda x: x.lower())
```

```python
new_df ['tags'] = new_df ['tags'].apply(lambda x: x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))                   #loving,loved,lover gets coverted to love
    return " ".join(y)      #convert list back to string
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') #cv is an object, max_features is number of words taken, remove stop words which are "and,of,to,from,are"
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector[0]
```

```python
cv.get_feature_names() #sort cv in alphabetical order
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)                #calculate the distance between each vector and the remaining all of the vectors
```

```python
similarity[0] #similarity between the first movie and the other 5000 movies
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]     #fetching index of the movie
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
